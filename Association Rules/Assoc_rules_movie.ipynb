# import necessary libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

# get data 

```python
movies = pd.read_csv('my_movies.csv')
movies
```

```python
movies = movies.iloc[:,5:]
movies
```

# creating frequent item list by using apriori algorithm

### frequent item_list with minimum support of 10 percentage

```python
freq_items_1 = apriori(movies,min_support=0.1,use_colnames=True)
freq_items_1
```

### frequent item_list with minimum support of 20 percentage

```python
freq_items_2 = apriori(movies,min_support=0.2,use_colnames=True)
freq_items_2
```

```python
freq_items_2.shape
```

### frequent item_list with minimum support of 30 percentage

```python
freq_items_3 = apriori(movies,min_support=0.3,use_colnames=True)
freq_items_3
```

```python
freq_items_3.shape
```

### frequent item_list with minimum support of 50 percentage

```python
freq_items_4 = apriori(movies,min_support=0.5,use_colnames=True)
freq_items_4
```

```python
freq_items_4.shape
```

# using association rules
### interpretation:taking a min_support eliminates the unnecessary itemsets

#### at min support of 10 percentage association rules :

```python
assc_rules_1 = association_rules(freq_items_1, metric='lift',min_threshold= 1)
assc_rules_1.sort_values('lift',ascending=False)[0:10]
strong_assoc_1 = assc_rules_1[assc_rules_1.lift>1][assc_rules_1.confidence>0.5]
strong_assoc_1
```

```python
strong_assoc_1.shape
```

#### at min support of 20 percentage association rules :

```python
assc_rules_2 = association_rules(freq_items_2, metric='lift',min_threshold= 1)
assc_rules_2.sort_values('lift',ascending=False)[0:10]
strong_assoc_2 = assc_rules_2[assc_rules_2.lift>1][assc_rules_2.confidence>0.5]
strong_assoc_2
```

```python
strong_assoc_2.shape
```

#### at min support of 30 percentage association rules :

```python
assc_rules_3 = association_rules(freq_items_3, metric='lift',min_threshold= 1)
assc_rules_3.sort_values('lift',ascending=False)[0:10]
strong_assoc_3 = assc_rules_3[assc_rules_3.lift>1][assc_rules_3.confidence>0.5]
strong_assoc_3
```

```python
strong_assoc_3.shape
```

#### at min support of 50 percentage association rules :

```python
assc_rules_4 = association_rules(freq_items_4, metric='lift',min_threshold= 1)
assc_rules_4.sort_values('lift',ascending=False)[0:10]
strong_assoc_4 = assc_rules_4[assc_rules_4.lift>1][assc_rules_4.confidence>0.5]
strong_assoc_4
```

### the people who watch sixth sense will definitely watch gladiator and people who watch patriot will not miss gladiator 

### at min support of 10 percent we have 135 associations with confidence greater than 50 percent and min lift threshold of 1; the associations may not be significant at this point but by increasing support gradually the strong asscociations were observed.
