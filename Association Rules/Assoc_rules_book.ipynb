# import necessary library

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

# getting data 

```python
book = pd.read_csv('book.csv')
book
```

# apriori algorithm

### taking min_support as 0.1 and forming a dataframe of frequency itemsets 

```python
freq_itemsets_1 = apriori(book,min_support=0.1,use_colnames=True)
freq_itemsets_1
```

```python
freq_itemsets_1.shape
```

### taking min_support as 0.2 and forming a dataframe of frequency itemsets 

```python
freq_itemsets_2 = apriori(book,min_support=0.2,use_colnames=True)
freq_itemsets_2
```

```python
freq_itemsets_2.shape
```

### taking min_support as 0.3 and forming a dataframe of frequency itemsets 

```python
freq_itemsets_3 = apriori(book,min_support=0.3,use_colnames=True)
freq_itemsets_3
```

```python
freq_itemsets_3.shape
```

### interpretation:taking a min_support eliminates the unnecessary itemsets

#### at min support of 10 percentage association rules :

```python
assoc_rules_1 = association_rules(freq_itemsets_1,metric='lift', min_threshold=1)
assoc_rules_1.sort_values('lift', ascending=False).head(10)
```

```python
dependance_1 =assoc_rules_1[assoc_rules_1.lift>1][assoc_rules_1.confidence>0.5]
dependance_1
```

```python
dependance_1.shape
```

#### at min support of 20 percentage association rules :

```python
assoc_rules_2 = association_rules(freq_itemsets_2,metric='lift', min_threshold=1)
assoc_rules_2.sort_values('lift', ascending=False).head(10)
```

```python
dependance_2 =assoc_rules_2[assoc_rules_2.lift>1][assoc_rules_2.confidence>0.5]
dependance_2
```

```python
dependance_2.shape
```

### the people who read childbooks will definitely read cookbooks

#### at min support of 10 percentage association rules :

```python
assoc_rules_3 = association_rules(freq_itemsets_3,metric='lift', min_threshold=1)
assoc_rules_3.sort_values('lift', ascending=False).head(10)
```

### at min support of 10 percent we have 49 associations with confidence greater than 50 percent and min lift threshold of 1 among the data where as at 30 percent support we dont have any associations.
